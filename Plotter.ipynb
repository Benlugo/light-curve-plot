```python
path = "C:\\Users\\Benlu\\documents\\FSRI_Research\\Star_Light_Curves"

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

count = 0

for filename in os.listdir(path):
    count = count + 1
    with open(path + "\\" + filename, 'r') as f:
        
        df = pd.read_csv(f)
        df.head()

        # finds the index in df of the lowest point
        maxIndex = 0
        for i in range(len(df)):
            if df["mag"][i] > df["mag"][maxIndex]:
                maxIndex = i

        #Now we shift the graph so that the lowest point
        #is our starting position
        shift = df["tdata"][maxIndex]
        df['tdata'] = (df['tdata']-shift)%1

        #creating our scatterplot
        plt.figure (figsize = (10,10))

        #We will rewrite tdata as phase and plot that as our x
        #The y axis will be the magnitude column
        sns.scatterplot (x= 'tdata', y ='mag', data = df)
        scatterplot = sns.scatterplot (x= 'tdata', y ='mag', data = df)

        #making the plot a bit more descriptive
        plt.title('Magitude vs. Phase of Star #' + str(count))
        plt.xlabel('Phase')
        plt.ylabel('Magnitude')
        yxis = plt.ylabel('Magnitude')
        scatterplot.invert_yaxis()
        plt.show()
```
